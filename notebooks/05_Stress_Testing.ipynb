# Agent 5: Macroeconomic Stress Testing

## Context
Regulatory bodies require financial institutions to undergo Stress Testing to prove they can survive severe economic shocks. The **Stress Testing Agent** simulates predefined crisis scenarios (e.g., market crashes, natural disasters) against the portfolio.

## Working Logic
1. **Scenario Definition:** Reads multipliers from `config.py` defining how different shocks affect Claims, Market Risk, and CAT Risk.
2. **Shock Application:** Applies these multipliers to the base portfolio metrics.
3. **Capital Consumption:** Calculates the capital required to survive the stressed claims environment.
4. **Solvency Ratio:** Computes the Stressed Solvency Ratio. If this drops below 100%, the portfolio triggers a "Capital Shortfall" warning.
5. **Product Vulnerability:** Analyzes which specific product lines (e.g., Motor, Property) are most vulnerable to the applied shocks.

## Key Concepts
- **Macroeconomic Shock:** Sudden, unexpected events that dramatically impact the economy.
- **Solvency Ratio:** Measures the size of a company's capital relative to all the risk it has taken. It demonstrates the company's ability to meet its long-term obligations.
